### Cleaning Data

1. mengapus tabel hr karena tidak diperlukan dan tidak relevan dengan pertanyaan bisnis

```python
del hour_df
```

2. menghapus kolom tabel yang tidak gunakan
menggunakan fitur drop pada kolom instant, dan winspeed karena tidak ada pertanyaan relevan dengan kolom ini

```python
drop_col = ['instant', 'windspeed']

for i in day_df.columns:
  if i in drop_col:
    day_df.drop(labels=i, axis=1, inplace=True)

day_df.head()
```

3. mengubah nama kolom agar mudah dibaca

```python
day_df.rename(columns={
    'dteday': 'dateday',
    'yr': 'year',
    'mnth': 'month',
    'weathersit': 'weather_cond',
    'cnt': 'count'
}, inplace=True)

day_df.head()
```

4. mengubah angka menjadi sebuah keterangan
mengubah angka menjadi keterangan agar lebih mudah dipahami saat membaca data

```python
day_df['month'] = day_df['month'].map({
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
})
day_df['season'] = day_df['season'].map({
    1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'
})
day_df['weekday'] = day_df['weekday'].map({
    0: 'Sun', 1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri', 6: 'Sat'
})
day_df['weather_cond'] = day_df['weather_cond'].map({
    1: 'Clear/Partly Cloudy',
    2: 'Misty/Cloudy',
    3: 'Light Snow/Rain',
    4: 'Severe Weather'
})
```

5. Mengubah tipe data dari object ke datetime dan ke category

```python
day_df['dateday'] = pd.to_datetime(day_df.dateday)
```

```python
day_df['season'] = day_df.season.astype('category')
day_df['year'] = day_df.year.astype('category')
day_df['month'] = day_df.month.astype('category')
day_df['holiday'] = day_df.holiday.astype('category')
day_df['weekday'] = day_df.weekday.astype('category')
day_df['workingday'] = day_df.workingday.astype('category')
day_df['weather_cond'] = day_df.weather_cond.astype('category')

day_df.head()
```

## Exploratory Data Analysis (EDA)

1. mengelompokkan pengguna sepeda baik casual dan registered berdasarkan bulan

```python
day_df.groupby(by='month').agg({
    'count': ['max', 'min', 'mean', 'sum']
})
```

berdasarkan hasil yang didapatkan, pada bulan juni merupakan bulan dengan pengguna terbanyak sedangkan pada bulan januari merupakan bulan dengan pengguna tersedikit

2. mengelompokkan pengguna baik casual dan registered berdasarkan cuaca

```python
day_df.groupby(by='weather_cond').agg({
    'count': ['max', 'min', 'mean', 'sum']
})
```

berdasarkan hasil yang didapatkan bahwa ketika cuaca sedang cerah "clear" atau sedikit mendung merupakan pengguna terbanyak, sedangkan saat hujan atau sedikit bersalju memiliki sedikit pengguna

3. mengelompokkan pengguna baik casual dan registered berdasarkan holiday

```python
day_df.groupby(by='holiday').agg({
    'count': ['max', 'min', 'mean', 'sum']
})
```

berdasarkan hasil yang didapatkan, pada hari liburan atau holiday tidak banyak pengguna, pengguna lebih memilih menyewa pada hari bukan liburan

4. perbandingan pengguna baik casual dan registered pada hari kerja vs weekend

```python
day_df.groupby(by='weekday').agg({
    'count':['max','min','mean']
})
```

berdasarkan hasil yang didapatkan, pengguna terbanyak pada hari jumat dan hari kamis dengan selisih 23. Sedangkan pengguna tersedikit pada hari minggu

5. mengelompokkan pengguna baik casual dan registered berdasarkan musim "season"

```python
day_df.groupby(by='season').agg({
    'casual': 'mean',
    'registered': 'mean',
    'count': ['max', 'min', 'mean']
})
```

berdasarkan hasil yang didapatkan, pengguna terbanyak pada musim fall(gugur), dan pengguna tersedikit pada musim spring (semi)

6. mengelompokkan suhu dan kelembapan berdasarkan musim

```python
day_df.groupby(by='season').agg({
    'temp': ['max', 'min', 'mean'],
    'atemp': ['max', 'min', 'mean'],
    'hum': ['max', 'min', 'mean']
})
```

berdasarkan hasil yang didapatkan, pada musim gugur (fall) memiliki suhu atau temperatur yang tinggi, sedangkan musim dingin (winter) memiliki kelembapan yang tinggi

7. korelasi antara variabel

## Visualization & Explanatory Analysis



### Pertanyaan 1: Bagaimana pengaruh cuaca terhadap jumlah pengguna sepeda?





```python
plt.figure(figsize=(8,6))
sns.barplot(
    x='weather_cond',
    y='count',
    data=day_df)

plt.title('Jumlah Pengguna Sepeda berdasarkan Cuaca')
plt.xlabel('Kondisi Cuaca')
plt.ylabel('Jumlah Pengguna Sepeda')
plt.show()
```

Hasil analisis:
*   Jumlah pengguna sepeda paling tinggi pada kondisi cuaca clear/partly cloudly
*   Jumlah pengguna sepeda paling rendah pada kondisi cuaca light snow/rain






### Pertanyaan 2: Bagaimana perbadingan jumlah pengguna sepeda pada tahun 2011 vs 2012?








```python
day_df['month'] = pd.Categorical(day_df['month'], categories=
    ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'],
    ordered=True)

monthly_counts = day_df.groupby(by=["month","year"]).agg({
    "count": "sum"
}).reset_index()

sns.lineplot(
    data=monthly_counts,
    x="month",
    y="count",
    hue="year",
    palette="rocket",
    marker="o")

plt.title("Jumlah pengguna sepeda berdasarkan Bulan dan tahun")
plt.xlabel(None)
plt.ylabel(None)
plt.legend(title="Tahun", loc="upper right")
plt.tight_layout()
plt.show()
```

pada visualisasi di atas terlihat bahwa:
*   Jumlah pengguna sepeda menunjukkan peningkatan 2x lipat di tahun 2012
*   jumlah pengguna tertinggi di tahun 2012(1) adalah pada bulan september, terendah pada bulan januari
* jumlah pengguna tertinggi di tahun 2011(0) adalah pada bulan juni, terendah pada bulan januari






### Pertanyaan 3: Apa pengaruh musim terhadap Jumlah Pengguna Sewa Sepeda baik Casual dan registered di tahun 2012?

```python
filtered_data_2012 = day_df[day_df["year"] == 1]

fig = px.bar(filtered_data_2012, x="season", y="count", title="Pengaruh Musim terhadap Jumlah Pengguna Sewa Sepeda pada tahun 2012")
fig.update_xaxes(title="musim")
fig.update_yaxes(title="Jumlah Pengguna Sewa Sepeda (cnt)")

fig.show()
```

Dari visualisasi di atas, dapat disimpulkan bahwa kondisi musim memiliki pengaruh yang signifikan terhadap jumlah pengguna sepeda.
Dimana pada musim gugur (fall) terdapat paling banyak pengguna sepeda yang diikuti oleh summer, winter, dan spring


### Pertanyaan 4: Apa pengaruh suhu atau temperatur serta kelembapan terhadap pengguna sepeda baik casual dan registered?

```python
plt.figure(figsize=(10, 6))

plt.subplot(1, 3, 1)
sns.scatterplot(
    x='temp',
    y='count',
    data=day_df,
    alpha=0.5
)
plt.title('Temperature vs Count')

plt.subplot(1, 3, 2)
sns.scatterplot(
    x='atemp',
    y='count',
    data=day_df,
    alpha=0.5
)
plt.title('Feels Like Temperature vs Count')

plt.subplot(1, 3, 3)
sns.scatterplot(
    x='hum',
    y='count',
    data=day_df,
    alpha=0.5
)
plt.title('Humidity vs Count')
```

berdasarkan hasil analisis:
- Hubungan antara temp dan count menunjukan korelasi yang positif.
- Hubungan antara atemp dan count juga menunjukan korelasi yang positif.
- Hubungan antara hum dan count menunjukan korelasi yang sedikit negatif, tetapi tidak terlalu signifikan dikarenakan persebarannya tidak terlalu condong ke kiri.

Sehingga dapat disimpulkan bahwa apabila temp dan atemp meningkat, maka jumlah pengguna sepeda juga cenderung akan meningkat. Akan tetapi, apabila humidity meningkat, belum tentu pengguna sepeda menurun namun, cenderung dapat menurunkan pengguna sepeda.
